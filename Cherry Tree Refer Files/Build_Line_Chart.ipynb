# Build Line Chart

````python
# Doing all the imports 

%matplotlib inline

import matplotlib.pyplot as plt

import pandas as pd #we can also plot data directly from a Pandas framework
````

### How a basic line Chart Looks like

````python
# Consider the following example:

time = [0, 1, 2, 3]

position = [0, 100, 200, 300]

plt.plot(time, position)

plt.xlabel('Time (hr)')

plt.ylabel('Position (km)')

plt.annotate('label', xy=(1.0,200), color='green')
````

### We will now read our data

````python
# Using pandas for importing
df_namerica = pd.read_csv('https://raw.githubusercontent.com/arpgpt/data-visualization/master/north_america.csv', index_col='country')
df_namerica
````

Before plotting, we convert the column headings from a string to integer data type, since they represent numerical values.

Extract year from last 4 characters of each column name.
The current column names are structured as 'gdpPercap_(year)',
so we want to keep the (year) part only for clarity when plotting GDP vs. years

To do this we use strip(), which removes from the string the characters stated in the argument
This method works on strings, so we call str before strip()

````python
years = df_namerica.columns.str.strip('gdpPercap_')
df_namerica.columns = years.astype(int) # Convert year values to integers, saving results back to dataframe
df_namerica
````

### We will now plot a line chart for all the countries in North America

````python
df_namerica.plot() #default plot 'kind' is line.
plt.ylabel('GDP per capita')
````

This plot looks confusing and crowded. Let us clean the data a bit more and plot a line chart after transposing the axis.



````python
df_namerica.T.plot() 
plt.ylabel('GDP per capita')
````

This line chart is much more descriptive in nature than the previous one. We can conclude that the United States has consistently better GDP per capita as compared to other countries on the list.

Let us now plot the data for the US in a separate line chart. For this, we will use loc.

````python
df_namerica.loc["United States"].plot(color = '#632cda') 

plt.ylabel('GDP per capita')
````

We can even plot line chart for different countries simultaneously.

````python
df_namerica.loc['Panama'].plot()  #Plot data for Panama, note that we use loc here and not iloc

df_namerica.loc['Dominican Republic'].plot()  #Plot data for Dominican Republic, note that we use loc here and not iloc

df_namerica.loc['Cuba'].plot()  #Plot data for Cuba, note that we use loc here and not iloc
````

Clearly, we are clueless as to which line plot belongs to which country. Hence, as a good coding hygiene one should always add legend to the plot to explain characters, symbols, or markings on a document that may be unfamiliar to the reader.

````python
df_namerica.loc['Panama'].plot()  #Plot data for Panama, note that we use loc here and not iloc

df_namerica.loc['Dominican Republic'].plot()  #Plot data for Dominican Republic, note that we use loc here and not iloc

df_namerica.loc['Cuba'].plot()  #Plot data for Cuba, note that we use loc here and not iloc

plt.legend()  #add legend()
````

##### Custom Line Color

To change the line color, use the color argument.

```python
df_namerica.loc["United States"].plot(color='green') 
plt.ylabel('GDP per capita')
```


````python
df_namerica.loc["United States"].plot(color='green') 
plt.ylabel('GDP per capita')
````

You can also add transparency to the color with the 

'alpha' argument 

(0=transparent, 1=opaque).

````python
df_namerica.loc["United States"].plot(color='green', alpha=0.4) 

plt.ylabel('GDP per capita')
````

#### Custom Line Style

You can choose between different line styles with the `linestyle` argument. A few of the line styles available are:

*   -
*   -- 
*   -.
*   : 



We will now see how these different line styles look like in a line plot.

##### Solid

````python
df_namerica.loc["United States"].plot(linestyle='-') #linestyle = 'solid'

plt.ylabel('GDP per capita')
````

##### dashed

````python
df_namerica.loc["United States"].plot(linestyle='--') #linestyle = 'dashed'

plt.ylabel('GDP per capita')
````

##### dashdhot

````python
df_namerica.loc["United States"].plot(linestyle='-.') #linestyle = 'dashdot'

plt.ylabel('GDP per capita')
````

##### dotted

````python
df_namerica.loc["United States"].plot(linestyle=':') #linestyle = 'dotted'

plt.ylabel('GDP per capita')
````

#### Custom Line Width

You can custom the line width usng the 'linewidth' argument.

````python
df_namerica.loc["United States"].plot(linewidth = 5) 

plt.ylabel('GDP per capita')
````

#### Add annotation in the line plot

````python
df_namerica.loc["United States"].plot(linewidth = 5) 

plt.ylabel('GDP per capita')

plt.annotate('GDP for USA', xy=(1970,30000), color='black') 

#here xy=() depicts the location of the annotate text
````

### Adding observations

Now we will add the observations on top of each point plotted in the line chart. Consider the following example where we plot values generated from two populated lists.

````python
df = df_namerica.loc["United States"]
df.plot(color = '#632cda' ,alpha= 0.5)
df
````

````python
xs = [1,2,3,4,5,6,7,8]

ys = [10,20,30,40,50,60,70,80]

plt.plot(xs,ys,'bo-',color = '#632dca' , alpha = 0.7)

for x,y in zip(xs,ys):

    label = (x,y)

    plt.annotate(label, # this is the text

                 (x,y), # this is the point to label

                 textcoords="offset points", # how to position the text

                 xytext=(0,10), # distance from text to points (x,y)

                 ha='center') # horizontal alignment can be left, right or center
````
